```python
# Librerías requeridas
import numpy as np
from matplotlib import pyplot as plt
from fuzzy import *
```

<h4>1. Definimos las variables linguistica de entrada y salida</h4>

```python
# Universos de discurso
# Entradas
SLI = np.linspace(0, 1023, 100)
SLC = np.linspace(0, 1023, 100)
SLD = np.linspace(0, 1023, 100)
ST = np.linspace(-256, 256, 100)
SOF = np.linspace(0, 55, 100)
# Salidas
MOV = np.linspace(0, 15000, 100)

# Funciones de pertenencia 
#SLI
MC1 = trapmf(SLI, [0, 0, 90, 240])
C1 = trapmf(SLI, [100, 240, 330, 485])
N1 = trapmf(SLI, [330, 485, 578, 760])
O1 = trapmf(SLI, [578, 760, 850, 997])
MO1 = trapmf(SLI, [850, 955, 1024, 1024])
#SLC
MC2 = trapmf(SLC, [0, 0, 90, 240])
C2 = trapmf(SLC, [100, 240, 330, 485])
N2 = trapmf(SLC, [330, 485, 578, 760])
O2 = trapmf(SLC, [578, 760, 850, 997])
MO2 = trapmf(SLC, [850, 955, 1024, 1024])
#SLD
MC3 = trapmf(SLD, [0, 0, 90, 240])
C3 = trapmf(SLD, [100, 240, 330, 485])
N3 = trapmf(SLD, [330, 485, 578, 760])
O3 = trapmf(SLD, [578, 760, 850, 997])
MO3 = trapmf(SLD, [850, 955, 1024, 1024])
#ST
R = trapmf(ST, [-256,-256,-150,-53])
IR = trapmf(ST, [-150,-53,53,150])
IN = trapmf(ST, [53,150,256,256])
#SOF
Ce = trimf(SOF, [0,13,27])
L = trimf(SOF, [25,41,55])
#MOV
P = trapmf(MOV,[0,0,900,1440])
AL = trimf(MOV,[900,2700,4500])
IL = trimf(MOV,[2700,4500,6300])
DL = trimf(MOV,[4500,6300,8100]) 
AR = trimf(MOV,[6300,8100,9900])
IR = trimf(MOV,[8100,9900,11700])
DR = trimf(MOV,[9900,11700,13500])
A = trapmf(MOV,[11700,13500,15000,15000])
```

```python
#gráficos para SLI
print("SLI")
plt.plot(SLI, MC1, label="MC")
plt.plot(SLI, C1, label="C")
plt.plot(SLI, N1, label="N")
plt.plot(SLI, O1, label="O")
plt.plot(SLI, MO1, label="MO")
plt.legend(loc='best')
plt.xlabel('Sensor Linea Izquierda')
plt.ylabel('$\mu (SLI)$')
plt.show()
#gráficos para SLC
print("SLC")
plt.plot(SLC, MC2, label="MC")
plt.plot(SLC, C2, label="C")
plt.plot(SLC, N2, label="N")
plt.plot(SLC, O2, label="O")
plt.plot(SLC, MO2, label="MO")
plt.legend(loc='best')
plt.xlabel('Sensor Linea Centrado')
plt.ylabel('$\mu (SLC)$')
plt.show()
#gráficos para SLD
print("SLD")
plt.plot(SLD, MC3, label="MC")
plt.plot(SLD, C3, label="C")
plt.plot(SLD, N3, label="N")
plt.plot(SLD, O3, label="O")
plt.plot(SLD, MO3, label="MO")
plt.legend(loc='best')
plt.xlabel('Sensor Linea Derecha')
plt.ylabel('$\mu (SLD)$')
plt.show()
#gráficos para ST
print("ST")
plt.plot(ST, R, label="R")
plt.plot(ST, IR, label="IR")
plt.plot(ST, IN, label="IN")
plt.legend(loc='best')
plt.xlabel('Sensor Terreno')
plt.ylabel('$\mu (ST)$')
plt.show()
#gráficos para SOF
print("SOF")
plt.plot(SOF, Ce, label="Ce")
plt.plot(SOF, L, label="L")
plt.legend(loc='best')
plt.xlabel('Sensor Obstaculo')
plt.ylabel('$\mu (SOF)$')
plt.show()
#gráficos para MOV
print("SLC")
plt.plot(MOV, P, label="P")
plt.plot(MOV, AL, label="AL")
plt.plot(MOV, IL, label="IL")
plt.plot(MOV, DL, label="DL")
plt.plot(MOV, AR, label="AR")
plt.plot(MOV, IR, label="IR")
plt.plot(MOV, DR, label="DR")
plt.plot(MOV, A, label="A")
plt.legend(loc='best')
plt.xlabel('MOVIMIENTO')
plt.ylabel('$\mu (MOV)$')
plt.show()
```

<h4>3. Método de Inferencia </h4>

```python
# Se calculan las funciones cortadas
M1 = cut(0.0, P)
M2 = cut(0.0,AL)
M3 = cut(0.0, IL)
M4 = cut(0.0, DL)
M5 = cut(0.226, AR)
M6 = cut(0.0, IR)
M7 = cut(0.0, DR)
M8 = cut(0.0, A)

# se unen
MOV_OUT = union([M1,M2,M3,M4,M5,M6,M7,M8])

#gráfico
print("SLC")
plt.plot(MOV, P, label="P")
plt.plot(MOV, AL, label="AL")
plt.plot(MOV, IL, label="IL")
plt.plot(MOV, DL, label="DL")
plt.plot(MOV, AR, label="AR")
plt.plot(MOV, IR, label="IR")
plt.plot(MOV, DR, label="DR")
plt.plot(MOV, A, label="A")
plt.plot(MOV, MOV_OUT, label="MOV_OUT", linewidth=5)
plt.legend(loc='best')
plt.xlabel('MOVIMIENTO')
plt.ylabel('$\mu (MOV)$')
plt.show()
```

<h4>4. Defuzzificación </h4>

```python
# Aplicamos defuzzificación
out_centroid = defuzz(MOV, MOV_OUT, 'centroid')
out_bisector = defuzz(MOV, MOV_OUT, 'bisector')
out_MOM = defuzz(MOV, MOV_OUT, 'MOM')
out_SOM = defuzz(MOV, MOV_OUT, 'SOM')
out_LOM = defuzz(MOV, MOV_OUT, 'LOM')
# Los valores son:
print ('out_centroid = ', out_centroid)
print ('out_bisector = ', out_bisector)
print ('out_MOM = ', out_MOM)
print ('out_SOM = ', out_SOM)
print ('out_LOM = ', out_LOM)
```

```python
#Graficamos para comparar
#Graficamos para visualizar sus posiciones

plt.plot(MOV, MOV_OUT, linewidth=2.5, linestyle="-", label="MOV_OUT")
plt.plot([out_centroid, out_centroid], [0, 1], linestyle=":", label="out_centroid")
plt.plot([out_bisector, out_bisector], [0, 1], linestyle=":", label="out_bisector")
plt.plot([out_MOM, out_MOM], [0, 1], linestyle=":", label="out_MOM")
plt.plot([out_SOM, out_SOM], [0, 1], linestyle=":", label="out_SOM")
plt.plot([out_LOM, out_LOM], [0, 1], linestyle=":", label="out_LOM")
plt.legend(loc='best')
plt.show()
```
